## Logistic Regression Practice - Part II (Data Cleaning)

#### Step 1 - Identify the variation of ages by classes

```python
# Identify outliers in the data with the .boxplot() method

plt.figure(figsize=(10,7))
sns.boxplot(x='Pclass', y='Age', data = train)
```

#### Step 2 - Create Function to fill in missing values


```python
# Create function to impute the age value if it is null

def impute_age(cols):
    Age = cols[0]
    Pclass = cols[1]
    
    if pd.isnull(Age):
        
        if Pclass == 1:
            return 37
        elif Pclass == 2:
            return 29
        else:
            return 24
    else:
        return Age
```

```python
# Apply function to impute the age for missing values
# The age column is at position 0
# The pclass column is at position 1
# axis -> argument refers to columns instead of rows to apply the impute_age function

train['Age'] = train[['Age', 'Pclass']].apply(impute_age, axis=1)
```

#### Step 3 - Visualize the updated dataset with the imputed age values


```python
# Create heatmap to display updated age values
# yticklabels -> argument removes the column labels
# cbar -> argument removes the color bar
# cmap -> argument defines the color of the map

sns.heatmap(train.isnull(), yticklabels=False, cbar=False, cmap='viridis')
```

#### Step 4 - Drop columns and rows that have too much missing data


```python
# Use the .drop() method to remove columns that have too much missing data

# axis -> argument referes to either row(0) or column(1) to specify
# inplace -> argument permanently removes the column instead of just removing the view

train.drop('Cabin', axis=1, inplace=True)
```

```python
train.head()
```

```python
# Use the .dropna() method to remove rows that contain missing values

train.dropna(inplace=True)
```

#### Step 5 - Create Dummy values for categorical data


```python
# Use the .get_dummies() method to convert categorical data into dummy values
# train['Sex'] refers to the column we want to convert
# drop_first -> argument avoids the multicollinearity problem, which can undermines
# the statistical significance of an independent variable.

sex = pd.get_dummies(train['Sex'], drop_first=True)
embark = pd.get_dummies(train['Embarked'], drop_first=True)
```

#### Step 6 - Create new dataframe with converted categorical data


```python
# Use  .concat() method to merge the series data into one dataframe


train = pd.concat([train, sex, embark], axis=1)
train.head()
```

#### Step 4 - Split data into features and target variables X and Y

```python
# Identify the column names again

df.columns
```

```python
# Split data into X (Features) and Y (Target) variables

X = df[['Avg. Area Income', 'Avg. Area House Age', 'Avg. Area Number of Rooms',
       'Avg. Area Number of Bedrooms', 'Area Population']]

y = df['Price']
```

#### Step 5 - Split data into training set and test sets with scikit learn


```python
# Import module to split data

from sklearn.model_selection import train_test_split
```

```python
# Review how the method works

# X - argument relates to the features of the dataset
# y - argument relates to the targets of the dataset
# test_size - argument is 40% of the dataset will be allocated to test set
# random_state - 

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=101)
```

#### Step 6 - Create and Train Model


```python
# Import module for Linear Regression

from sklearn.linear_model import LinearRegression
```

```python
# Create a LinearRegression object to train data (i.e. called instantiate)

lm = LinearRegression()
```

```python
# Fit (i.e. Train) the model on the training dataset

lm.fit(X_train, y_train)
```

#### Step 7 - Evaluate the Model created


```python
# Identify the intercept

print(lm.intercept_)
```

```python
# Identify the coefficients of the training dataset features

lm.coef_
```

```python
# Create dataframe to evaluate coefficients

cdf = pd.DataFrame(lm.coef_,X.columns, columns=['Coeff'])
cdf
```

### Linear Regression - Part II

### Agenda

* Obtain predictions from our model

### Predictions

#### Step 1 - make predictions

```python
# Use the .predict() method to obtain predictions from the model

predictions = lm.predict(X_test)
predictions
```

```python
# Remember we have the y_test variable that holds the correct labeled values
# The question is, how close is the prediction to the actual values

y_test
```

#### Step 2 - Evaluate predictions against correct y_test values

```python
# Use scatter plot to evaluate how well the predictions line up with the actuals

plt.scatter(y_test, predictions, edgecolors='w')
```

```python
# Use the seaborn .distplot() to understand the variance of the residuals

sns.distplot((y_test-predictions))
```

#### Step 3 - Use loss functions to understand variance between what's predicted and the actual value

```python
# Import the metrics module

from sklearn import metrics
```

```python
# Use the Mean Absolute Error (MAE) - Easiest

metrics.mean_absolute_error(y_test, predictions)
```

```python
# Use the Mean Squared Error (MSE) - Punishes larger errors...More useful in real world applications

metrics.mean_squared_error(y_test, predictions)
```

```python
# Use the Root Mean Squared Error (RMSE) - More popular b/c it is interpretable to the 'y' labels

# We need to use numpy here to obtain the root of the MSE

np.sqrt(metrics.mean_squared_error(y_test, predictions))
```

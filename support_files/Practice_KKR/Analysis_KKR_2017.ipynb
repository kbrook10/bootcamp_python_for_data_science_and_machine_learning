### (2) How many hours were recorded during the weekdays

```python
# Create a series of weekdays
# for each entry using .dt.weekday (note: 0 starts as Monday)
weekdays = df['Date'].dt.weekday
```

```python
# Use the .assign() method to create a new weekday column; This takes a copy of the df instead of a view
df = df.assign(weekdays = weekdays)
df[['Date','weekdays']].tail(9)
```

```python
# Now we need to replace the integers with actual weekdays
wk_day_names = ['Mon','Tue','Wed','Thu','Fri','Sat','Sun']
wk_day_dict = {key: wk_day_names[key] for key in range(7)}
wk_day_dict
```

```python
# Use .apply() method to assign a custom function
#def day_of_week(idx):
#    return wk_day_dict[idx]

df['weekdays'] = df['weekdays'].apply(lambda idx:wk_day_dict[idx])
```

### (3) How do we group by the weekdays

```python
# Use the .groupby() method to group the weekdays and then sum the recorded hours
wk_day_totals = df.groupby('weekdays').sum()
wk_day_totals = wk_day_totals.loc[wk_day_names]
wk_day_totals[['Hours','Hours Rounded']]
#wk_day_totals.sort_values(by='Hours Rounded', ascending=False)
```

```python
# Use the .groupby() method to group the weekdays and then obtain the count the recorded hours
wk_day_count = df.groupby('weekdays').count()
wk_day_count = wk_day_count.loc[wk_day_names]
wk_day_count[['Hours','Hours Rounded']]
#wk_day_count[['Hours','Hours Rounded']].sort_values(by='Hours Rounded', ascending=False)
```

```python
# Time to visualize the Sums and the Counts Data
# Use the .plot() method to display a bar graph for the wk_day_count data
wk_day_count['Hours Rounded'].plot(kind='bar')
```

### Task with the most recorded hours?

```python
# Conduct aggregate function based on the tasks label
task_summary = df.groupby('Task').sum().sort_values('Hours Rounded', ascending=False)
task_summary
```

```python
# What is the Pareto analysis for the Task Hours
total_hours = df['Hours Rounded'].sum()

# Add a new column to divide for percentages
task_summary['percent_total_hours'] = task_summary['Hours Rounded'] / total_hours 
```

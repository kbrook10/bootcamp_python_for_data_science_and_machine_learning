## Seaborn - Grids

### Step (1) - Setup

```python
# import necessary library, configure matplotlib plotting and load in data set
import seaborn as sns; sns.set()
import matplotlib.pyplot as plt
%matplotlib inline
iris = sns.load_dataset('iris')
iris.head(10)
```

### Step (2) - Explore Dataset

```python
# Identify the unique values of a column within the dataset
iris['species'].value_counts()
```

```python
# Create Cluster map with .corr() method
imx = iris.corr()
sns.clustermap(imx, cmap='coolwarm')
```

```python
# Conduct a pairplot on the data to identify similarities
sns.pairplot(iris, hue='species', diag_kind='kde')
```

### Step (3) - How to configure and use PairGrid to further customize pairplots


```python
# Use the .PairGrid() method to generate blank plots of the numerical columns within the dataset
g = sns.PairGrid(iris)
```

```python
# Use the .PairGrid() method to generate blank plots of the numerical columns within the dataset
g = sns.PairGrid(iris)

# Use the .map() method to pass in the plot type desired -> These (2) lines of code need to be in same cell
g.map(plt.scatter)
```

```python
# Use the .PairGrid() method to generate blank plots of the numerical columns within the dataset
g = sns.PairGrid(iris, hue = 'species')

# Use the .map() method to pass in desired plot type
# .map_diag -> references grids on the diagonal
# .map_upper -> references grids in the upper right
# .map_lower -> references grids in the lower left
# sns.distplot -> references a distribution plot with kde
# plt.scatter -> refernces a scatter plot
# sns.kdeplot -> references 2d kernel density estimation plot

g.map_diag(sns.distplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

### Step (4) - How to use Facet Grid


```python
# Load in dataset
tips = sns.load_dataset('tips')
tips.head(5)
```

```python
# Use the .FacetGrid method to create blank grids; think or it link creating a blank object
g = sns.FacetGrid(data=tips,col='time',row='smoker')
```

```python
# Use the .FacetGrid method to create blank grids; think or it link creating a blank object
g = sns.FacetGrid(data=tips,col='time',row='smoker')

# Use the .map() function to define the plot time over the desired column
g.map(sns.distplot,'total_bill')
```

```python
# Use the .FacetGrid method to create blank grids; think or it link creating a blank object
g = sns.FacetGrid(data=tips,col='time',row='smoker')

# Use the .map() function to define the plot time over the desired column
g.map(plt.scatter,'total_bill','tip')
```

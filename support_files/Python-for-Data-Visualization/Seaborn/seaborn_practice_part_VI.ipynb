## Seaborn - How to use Style and Colors

### Step (1) Setup initial environment and dataset

```python
# Import necessary libraries, load dataset and configure matplotlib to display plots inline
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head(10)
```

### Step (2) How to adjust styles

```python
# Use the .countplot() method to create a simple histogram
# Use the .set_style() argument to pass in styling to the plot

sns.set_style('whitegrid')
sns.countplot(x='sex', data=tips)
```

```python
# Use the .countplot() method to create a simple histogram
# Use the .set_style() argument to pass in styling to the plot
# Use the .despine() function to remove the top and right spines of the plot by default


sns.set_style('ticks')
sns.countplot(x='sex', data=tips)
sns.despine()
```

```python
# Use the .countplot() method to create a simple histogram
# Use the .set_style() argument to pass in styling to the plot
# Use the .despine() function to remove specific spines of the plot by passing in arguments


sns.set_style('ticks')
sns.countplot(x='sex', data=tips)
sns.despine(left=True, bottom=True)
```

### Step (3) How to adjust size and aspect ratios of plot

```python
# Use the .countplot() method to create a simple histogram
# Use the callback .figure() method of matplotlib with the argument figsize to adjust the size...
# figsize argument requires a tuple -> i.e. an immutable list


plt.figure(figsize=(12,3))
sns.countplot(x='sex', data=tips)
```

### Step (4) How to adjust scale and context

```python
# Use the .countplot() method to create a simple histogram
# Use the callback .figure() method of matplotlib with the argument figsize to adjust the size...
# .set_context() method provides scale based on the type of presentation (paper, notebook, talk, poster)


sns.set_context(context='poster', font_scale=1.5)
sns.countplot(x='sex', data=tips)
```

### Step (5) How to adjust pallets and colors

```python
# Use the .lmplot() method to generate a scatter plot with linear regression on top
# x references the x-axis column value
# y references the y-axis column value
# data references the dataset being used
# hue references a further variable to divide the data
# palette references the color string for the plot

sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', palette='seismic')
```

## Seaborn Practice Part II - Categorial Plots

### Step (1) Import necessary libraries

```python
# Import the necessary libraries, dataset and configure matplotlib to run within the notebook
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head(10)
```

### How to use the .barplot() method to create bar plots

```python
# Use the .barplot() method with arguments x, y and data to generate a bar plot
# x references the categorical data
# y references the numerical data
# data references the dataset

sns.barplot(x='sex', y='total_bill', data=tips)
```

```python
# Import numpy to use the standard deviation method
import numpy as np
```

```python
# Modify the estimator argument within the .barplot() method
sns.barplot(x='sex', y='total_bill', data=tips, estimator=np.std)
```

### How to use the .countplot() method to count occurrences

```python
# Use the .countplot() method with the x (categorical) and dataset arugments
sns.countplot(x='sex', data=tips)
```

### How to use the .boxplot() method to display distribution of categorical data

```python
# Use the .boxplot method with x(categorical data), y(numerical data) and data (dataset) arguments
sns.boxplot(x='day', y='total_bill', data=tips)
```

```python
# Use the hue argument to pass in another category to further split the data
sns.boxplot(x='day', y='total_bill', data=tips, hue='smoker')
```

### How to use the .violinplot() method to display distribution of categorical data


```python
# Use the .violinplot() method with arguments x (categorical), y(numerical) and data(dataset) arguments
sns.violinplot(x='day', y='total_bill', data=tips)
```

```python
# Use the .violinplot() method with additional arguments hue for additional categories and split to dist. hue
sns.violinplot(x='day', y='total_bill', data=tips, hue='sex', split=True)
```

### How to use the .stripplot() method to display scatter plot of categorical data


```python
### How to use the .stripplot() method to display scatter plot of categorical data
# x references categorical data
# y referenses numerical data
# data references dataset

sns.stripplot(x='day', y='total_bill', data=tips)
```

```python
### How to use the .stripplot() method to display scatter plot of categorical data
# x references categorical data
# y referenses numerical data
# data references dataset
# jitter reduces the noise and spaces out the points

sns.stripplot(x='day', y='total_bill', data=tips, jitter=True)
```

```python
### How to use the .stripplot() method to display scatter plot of categorical data
# x references categorical data
# y referenses numerical data
# data references dataset
# jitter reduces the noise and spaces out the points
# hue provides additional segmentation of the data
# split divides the data

sns.stripplot(x='day', y='total_bill', data=tips, jitter=True, hue='sex', dodge=True)
```

### How to use the .swarmplot() method to display scatter plot distribution of categorical data


```python
### How to use the .swarmplot() method to display scatter plot distribution of categorical data
# x references categorical data
# y referenses numerical data
# data references dataset

sns.swarmplot(x='day', y='total_bill', data=tips)
```

```python
### How to use the .swarmplot() method with the violin plot to display scatter plot distribution of categorical data
# x references categorical data
# y referenses numerical data
# data references dataset

sns.violinplot(x='day', y='total_bill', data=tips)
sns.swarmplot(x='day', y='total_bill', data=tips, color='black')
```

### How to use the .factorplot() method to bar plot of categorical data


```python
### How to use the .factorplot() method to display bar plot of categorical data
# x references categorical data
# y referenses numerical data
# data references dataset
# kind references the type of plot

sns.factorplot(x='day', y='total_bill', data=tips, kind='bar')
```

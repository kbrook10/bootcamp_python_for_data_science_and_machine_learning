## Seaborn - How to use Regression Plots

### Step (1) Setup initial environment and dataset

```python
# Import necessary libraries, load dataset and configure matplotlib to display plots inline
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head(10)
```

### Step (2) How to use the .lmplot() method for regression plots

```python
# Use the .lmplot() method with arguments x (x axis elements), y (y axis elements), and data (data points) to 
# generate a scatter plot with a linear regression plot on top

sns.lmplot(x='total_bill', y='tip', data=tips)
```

```python
# Use the .lmplot() method with arguments x (x axis elements), y (y axis elements), and data (data points) to 
# generate a scatter plot with a linear regression plot on top

# Use hue argument to reference further division of the data

sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex')
```

```python
# Use the .lmplot() method with arguments x (x axis elements), y (y axis elements), and data (data points) to 
# generate a scatter plot with a linear regression plot on top

# Use hue argument to reference further division of the data
# Use the marker argument to reference matplotlib tip markers for further differentiation

sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', markers=['o','v'])
```

```python
# Use the .lmplot() method with arguments x (x axis elements), y (y axis elements), and data (data points) to 
# generate a scatter plot with a linear regression plot on top

# Use hue argument to reference further division of the data
# Use the marker argument to reference matplotlib tip markers for further differentiation
# Use the scatter_kws argument to access matplotlib under the hood to adjust the marker size

sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', markers=['o','v'],
          scatter_kws={'s':200})
```

```python
# Use the .lmplot() method with arguments x (x axis elements), y (y axis elements), and data (data points) to 
# generate a scatter plot with a linear regression plot on top

# Use col argument to create separate plots by category, similar to the grid plots

sns.lmplot(x='total_bill', y='tip', data=tips, col='sex')
```

```python
# Use the .lmplot() method with arguments x (x axis elements), y (y axis elements), and data (data points) to 
# generate a scatter plot with a linear regression plot on top

# Use col argument to create separate plots by category, similar to the grid plots
# Use row argument to futher separate plots by category

sns.lmplot(x='total_bill', y='tip', data=tips, col='sex', row='time')
```

### Step (3) How to use the aspect argument within the .lmplot() method to alter the size and aspect ratio


```python
# Use the .lmplot() method with arguments x (x axis elements), y (y axis elements), and data (data points) to 
# generate a scatter plot with a linear regression plot on top

# Use col argument to create separate plots by category, similar to the grid plots
# Use row argument to futher separate plots by category
# Use the aspect argument to alter the aspect of the plot

sns.lmplot(x='total_bill', y='tip', data=tips, col='sex', hue='sex',
          aspect=0.6)
```

```python
# Use the .lmplot() method with arguments x (x axis elements), y (y axis elements), and data (data points) to 
# generate a scatter plot with a linear regression plot on top

# Use col argument to create separate plots by category, similar to the grid plots
# Use row argument to futher separate plots by category
# Use the aspect argument to alter the aspect of the plot
# Use the size argument to alter the size of the plot


sns.lmplot(x='total_bill', y='tip', data=tips, col='sex', hue='sex',
          aspect=0.6, size = 8)
```

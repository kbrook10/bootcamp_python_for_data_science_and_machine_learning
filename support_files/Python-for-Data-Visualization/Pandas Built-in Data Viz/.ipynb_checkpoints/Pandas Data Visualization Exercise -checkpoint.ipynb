# Pandas Data Visualization Exercise

This is just a quick exercise for you to review the various plots we showed earlier. Use **df3** to replicate the following plots. 

```python
import pandas as pd
import matplotlib.pyplot as plt
df3 = pd.read_csv('df3')
%matplotlib inline
```

```python
df3.info()
```

```python
df3.head()
```

** Recreate this scatter plot of b vs a. Note the color and size of the points. Also note the figure size. See if you can figure out how to stretch it in a similar fashion. Remeber back to your matplotlib lecture...**

```python
# Use the .plot.scatter() method to generate plot with arguments a (x-axis) and b (y-axis)
# Use column a values for the x axis
# Use column b values for the y axis
# Use the c argument to specify the color
# Use the edgecolor argument to create border of back around data points
# Use the s argument to adjust the size of the data points
# Use figsize argument to alter the size of the plot

#df3.axis([-0.2, 1.2, -0.2, 1.2])
df3.plot.scatter(x='a', y='b',c='red',edgecolor='k', s=40, figsize=(12,3))
```

** Create a histogram of the 'a' column.**

```python
# Use the .plot.hist() method to generate a histogram
# Use edgecolor argument for borders on bars

df3['a'].plot.hist(edgecolor='k')
```

** These plots are okay, but they don't look very polished. Use style sheets to set the style to 'ggplot' and redo the histogram from above. Also figure out how to add more bins to it.***

```python
# Use the .style.use() method to alter to look of the histogram
import matplotlib.pyplot as plt


plt.style.use('ggplot')
df3['a'].plot(kind='hist', bins=25, edgecolor='w', alpha=0.5)
```

** Create a boxplot comparing the a and b columns.**

```python
# Create boxplot with arguments a and b

df3.plot.box(x=['c','d'])
```

** Create a kde plot of the 'd' column **

```python
# Use kde plot method
# Use c for the color adjustment
# Use alpha to increase transparency

df3['d'].plot.kde(c='red', alpha=.5)
```

** Figure out how to increase the linewidth and make the linestyle dashed. (Note: You would usually not dash a kde plot line)**

```python
# Use kde plot method
# Use c for the color adjustment
# Use lw for the line width
# Use ls for the line style

df3['d'].plot.kde(c='red', lw=3, ls='--')
```

** Create an area plot of all the columns for just the rows up to 30. (hint: use .ix).**

```python
# Use the .plot.area() method 
# Use cmap to change the color
# Use .ix to slice the dataframe to the first 30 rows
# Use alpha to increase the transparency

df3.ix[0:30].plot.area(cmap='coolwarm',alpha=0.6)
```

## Bonus Challenge!
Note, you may find this really hard, reference the solutions if you can't figure it out!
** Notice how the legend in our previous figure overlapped some of actual diagram. Can you figure out how to display the legend outside of the plot as shown below?**

** Try searching Google for a good stackoverflow link on this topic. If you can't find it on your own - [use this one for a hint.](http://stackoverflow.com/questions/23556153/how-to-put-legend-outside-the-plot-with-pandas)**

```python
# Use the .plot.area() method 
# Use cmap to change the color
# Use .ix to slice the dataframe to the first 30 rows
# Use alpha to increase the transparency
# Use .legend() method to adjust placement of legend

df3.ix[0:30].plot.area(cmap='coolwarm',alpha=0.6).legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

## Pandas - Data Visualization - Built In Features

### Step (1) Setup initial environment

```python
# Import the necessary dependencies and configure matplotlib to display inline
import numpy as np
import pandas as pd
%matplotlib inline
pd.__version__
```

```python
# Load in the test datasets
# Use the pd.read_csv() method to load in dataframe
# Use the index_col argument to specify how to generate he index
# *Note: this has a timeseries value and that is why we allocated the index to that

df1 = pd.read_csv('df1', index_col=0)
df1.head(10)
```

```python
# Load in the test datasets
# Use the pd.read_csv() method to load in dataframe
# *Note: This data just has a sequential index

df2 = pd.read_csv('df2')
df2.head(10)
```

### Step (2) How to generate Historgrams

```python
# Use the .hist() method on the dataframe to generate a histogram
df1['A'].hist(bins=30)
```

```python
# ** Method A **

# Using the plot() method and the kind argument with different values for plots
# Use the kind argument to define the type of plot
# Use the bins argument to specify the quantity of bins
# Use the edgecolor argument when using jupyter notebooks if the bins do not display the edges.

df1['A'].plot(kind='hist', edgecolor="k", bins=30)
```

```python
# ** Method B **

# Using the plot() method and the kind argument with different values for plots
# Use the bins argument to specify the quantity of bins
# Use the edgecolor argument when using jupyter notebooks if the bins do not display the edges.

df1['A'].plot.hist(edgecolor="k", bins=30)
```

### Step (3) How to generate Area Plot

```python
# Use the .area() method to generate area plot
# Use alpha to apply transparency to the plot

df2.plot.area(alpha=0.4)
```

### Step (4) How to generate Bar Plot

```python
# Use the .bar() method to generate a bar plot

df2.plot.bar()
```

```python
# Use the .bar() method to generate a bar plot
# Use the stacked argument to stack the bars

df2.plot.bar(stacked=True)
```

### Step (5) How to generate Line Plot

```python
# Use the .line() method to plot
# Use x argument to reference the desired x-axis value
# Use y argument to reference the desired y-axis value
# We can also use other matplotlib agruments: figsize, lw, etc.

df1.plot.line(x=df1.index, y= 'B', figsize=(12,3), lw=1)
```

### Step (6) How to generate Scatter Plot

```python
# Use the .scatter method with x and y string arguments
# A references the x-axis argument
# B references the y-axis argument


df1.plot.scatter(x='A', y='B')
```

```python
# Use the .scatter method with x and y string arguments
# A references the x-axis argument
# B references the y-axis argument
# c refernces the color

df1.plot.scatter(x='A', y='B', c='C')
```

```python
# Use the .scatter method with x and y string arguments
# A references the x-axis argument
# B references the y-axis argument
# c refernces the color
# cmap argument alters the color of the plot

df1.plot.scatter(x='A', y='B', c='C', cmap='coolwarm')
```

```python
# Use the .scatter method with x and y string arguments
# A references the x-axis argument
# B references the y-axis argument
# s refernces the size
# cmap argument alters the color of the plot

df1.plot.scatter(x='A', y='B', s=(df1['C']*72), edgecolor='k')
```

### Step (7) How to generate Box Plot

```python
# Use the .box() method
df2.plot.box()
```

### Step (8) How to generate plot for Bi-variate (i.e. two variables) data

```python
# Setup data
df = pd.DataFrame(np.random.randn(1000,2), columns=['a','b'])
df.head()
```

```python
# Use .hexbin() method
# Use gridsize argument to increase the size
# Use cmap argument to change the default color

df.plot.hexbin(x='a', y='b', gridsize=25, cmap='coolwarm')
```

### Step (9) How to generate KDE Plots

```python
# Option A - Use .kde() method
df2['a'].plot.kde()
```

```python
# Option B - Use .density() method
df2['a'].plot.density()
```

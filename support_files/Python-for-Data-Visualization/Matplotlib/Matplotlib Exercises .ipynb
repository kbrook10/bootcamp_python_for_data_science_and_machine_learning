## Exercise 1

** Follow along with these steps: **
* ** Create a figure object called fig using plt.figure() **
* ** Use add_axes to add an axis to the figure canvas at [0,0,1,1]. Call this new axis ax. **
* ** Plot (x,y) on that axes and set the labels and titles to match the plot below:**

```python
# Process
# Step (1) use the .figure() method to create figure object
fig = plt.figure()
# Step (2) use the .add_axes() method to create the axes
ax = fig.add_axes([0,0,1,1])
# Step (3) use the .plot() method to plot the axes
ax.plot(x,y)
# Step (4) add the x, y labels and title
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('title')

```

## Exercise 4

** Use plt.subplots(nrows=1, ncols=2) to create the plot below.**

```python
# Process
# Step (1) Create figure with axes of rows 1 and columns 2, using the .subplots() method
fig, axes = plt.subplots(nrows=1, ncols=2)
```

** Now plot (x,y) and (x,z) on the axes. Play around with the linewidth and style**

```python
# Process
# Step (1) Create figure with axes of rows 1 and columns 2, using the .subplots() method
fig, axes = plt.subplots(nrows=1, ncols=2)

# Step (2) Plot the axes and adjust the linewidth and styles on the .plot() method arguments
axes[0].plot(x,y,color='blue',lw=2.5,ls='--')
axes[1].plot(x,z,color='red',lw=2.5,ls='-')
```

** See if you can resize the plot by adding the figsize() argument in plt.subplots() are copying and pasting your previous code.**

```python
# Process
# Step (1) Create figure with axes of rows 1 and columns 2, using the .subplots() method
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12,2))

# Step (2) Plot the axes and adjust the linewidth and styles on the .plot() method arguments
axes[0].plot(x,y,color='blue',lw=2.5,ls='-')
axes[0].set_xlabel('x')
axes[0].set_ylabel('y')
axes[1].plot(x,z,color='red',lw=2.5,ls='--')
axes[1].set_xlabel('x')
axes[1].set_ylabel('z')
```
